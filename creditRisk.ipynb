# Str into num type

```python
loan_data['int_emp_lenght'] = loan_data['emp_length'].str.replace('\+ years', '')
loan_data['int_emp_lenght']= loan_data['int_emp_lenght'].str.replace('< 1 year',str(0))
loan_data['int_emp_lenght'] = loan_data['int_emp_lenght'].str.replace('nan',str(0))
loan_data['int_emp_lenght']  = loan_data['int_emp_lenght'].str.replace('years','')
loan_data['int_emp_lenght'] = loan_data['int_emp_lenght'].str.replace('year','')
```

```python
type(loan_data['int_emp_lenght'][0])
```

```python
loan_data['int_emp_lenght'] = pd.to_numeric(loan_data['int_emp_lenght'])
```
